```python
# Import needed packages

#import active_subspaces as ac
import numpy as np
import scipy as sp
from scipy.optimize import minimize
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def misfit_1(x):
    return 0.5*(4*(2*x[0]-x[1]-0.1)**2 + (2/9)*(x[0]-0.1)**2 + (16/9)*(x[0]-0.1)*(x[1]-0.2) + (32/9)*(x[1]-0.2)**2)
```

```python
sp.optimize.minimize(misfit_1,x0=np.array((1,0.5)))
```

```python
def nonlin(x):
    return x**5 + x**3 +x

def nonlinp(x):
    return 5*x**4 + 3*x**2 +1

def lin(x):
    return nonlin(1)+ nonlinp(1)*(x-1)
```

```python
N=100
samps=np.array(0)
co=0

while co<N:
    mu1, sigma1 = 1, 0.5 # mean and standard deviation
    s = np.random.normal(mu1, sigma1)
    if 3.99 < nonlin(s) < 4.01:
        samps = np.vstack((samps,s))
        co=co+1
    else:
        samps=samps
        co=co
        
samps=samps[1:]
```

```python
js=samps[1]
js[0]
```

```python
def Tfun(x):
    return 0.5*(4*(nonlin(x)-4)**2+2*(x-1)**2)
```

```python
def Sfun(x):
    return 0.5*(4*(lin(x)-4)**2+2*(x-1)**2-(1/9)*(lin(x)-3)**2)
```

```python
Tsamps=np.zeros(N)

for i in range(0,N):
    Tsamps[i]=Tfun(samps[i][0])
```

```python
ind=np.argmin(Tsamps)

Tstar=samps[ind][0]

print(Tstar)

Tfun(Tstar)
```

```python
def thisprior(x):
    return np.exp(-Sfun(x))
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.01)

# red dashes
plt.plot(t, thisprior(t), 'r--')
plt.show()
```

```python

# red dashes
plt.plot(samps, nonlin(samps), 'r--')
plt.show()
```

```python
N=1000000

mu1, sigma1 = 1, 0.5 # mean and standard deviation
samps2 = np.random.normal(mu1, sigma1, N)

samps2[3]
```

```python


Tsamps2=np.zeros(N)

for i in range(0,N):
    Tsamps2[i]=nonlin(samps2[i])
```

```python
kdesol=sp.stats.gaussian_kde(Tsamps2)
```

```python
def pushfwd(x):

    return kdesol(x)[0]
```

```python
def troy(x):
    
    return (1/(2*np.pi))*np.exp(-(x-1)**2)*np.exp(-2*(nonlin(x)-4)**2)/pushfwd(x)
```

```python
def troyfm(x):
    return x*troy(x)
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(-1, 5, 0.1)

# red dashes
plt.plot(t, troy(t), 'r--', t, thisprior(t), 'b--')
plt.show()
```

```python
sp.integrate.quad(troy, -2, 3)
```

```python
def fun(x):
    return 2*x
```

```python
# Solve fwd problem

M=1000

# Draws in Lambda
sampz=np.random.normal(0.1,np.sqrt(0.5),M)

# Push draws through f

fsampz=np.zeros(M)

for i in range(0,M):
    fsampz[i]=fun(sampz[i])
    
```

```python
kdesol1=sp.stats.gaussian_kde(fsampz)
```

```python
def prior1(x):
    
    return (1/np.sqrt(np.pi))*np.exp(-(x-0.1)**2)
```

```python
# The pushfwd can also be calculated in this case!

# This is now pushfwd of troya!

def pushfwd1(x):
    
    return (1/np.sqrt(0.5*np.pi))*np.exp(-2*(x-0.25)**2)

# This is the pushfwd of statistical bayesian!

def pushfwd2(x):

    return (1/np.sqrt(2*np.pi*0.1986))*np.exp(-(x-(6/25))**2/(2*0.1986))
```

```python
def obs1(x):
    return (np.sqrt(2)/np.sqrt(np.pi))*np.exp(-2*(x-0.25)**2)
```

```python
def troya(x):
    
    return (1/np.sqrt(np.pi/8))*np.exp(-8*(x-0.125)**2)
```

```python
def bayes1(x):
    
    return (1/np.sqrt(2*np.pi*0.04966))*np.exp(-(x-(3/25))**2/(2*0.04966))
```

```python
# You can't evaluate a PDF for a probability! Naughty!!!!

troya(0.1)
```

```python
sp.integrate.quad(bayes1, -2, 2)
```

```python
# Parameter plots

# evenly sampled time at 200ms intervals
t = np.arange(-3, 3, 0.01)

# red dashes
plt.plot(t, prior1(t), 'b--', t, troya(t), 'g--', t, bayes1(t), 'r--')


plt.axis([-2, 3, 0, np.max(bayes1(t))+0.25])
plt.xlabel('$\lambda$')
plt.ylabel('')
plt.title('Distributions in Parameter Space')

plt.legend(('$\pi_{\Lambda}^{prior}(\lambda)$', '$\pi_{\Lambda}^{update}(\lambda)$ obtained with DCI', '$\pi_{\Lambda}^{update}(\lambda)$ obtained with Statistical Bayes'),
           loc='upper right')

#plt.xticks(np.arange(-1, 1, 0.125))

params = {'legend.fontsize': 35,
          'legend.handlelength': 3}


plt.rcParams["figure.figsize"] = (35,25)

plt.rcParams['figure.dpi'] = 80
plt.rcParams['savefig.dpi'] = 100

plt.rcParams['font.size'] = 22
plt.rcParams['figure.titlesize'] = 'xx-large'
plt.rcParams.update(params)

plt.show()
```

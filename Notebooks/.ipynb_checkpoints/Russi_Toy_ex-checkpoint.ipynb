# Comparing Active Subspace Analysis with Principal Component Analysis


Here, we attempt to show the similarities and differences in using Active Subpaces and PCA to perform dimension reduction.

For PCA, we use a canned $\texttt{scikit}$ algorithm, which will use a full SVD for the problem we explore.

For the active subspace analysis, we mimic Paul Constantine's "basic" tutorial in his active subspaces library on GitHub, using a different toy function. All text directly copied from Paul's original notebook are cited with quotations throughout this notebook.

We will use the toy function from TM Russi, Section 6.6:

$$f: \mathbb{R}^{30} \to \mathbb{R}$$

$$f: \vec{x} \to f(\vec{x})$$

$$
f(\vec{x}) = \sin(x_1) + x_2 x_3 x_5 + x_1x_4^2 +e^{x_3}\cos(x_5^3) 
$$

Notice we have defined $f$ to have no depedence on $x_6,\ldots,x_{30}$; to model those parameters as useless predictors, we will add a random noise factor $\epsilon$ to all function outputs, $\tilde{f}(\vec{x})=f(\vec{x})+\epsilon$. The random noise factor $\epsilon$ is scaled to the unperturbed function output; in particular, we'll let $\epsilon \propto \left|\left|f(\vec{x})\right|\right|$. We begin by assuming each varibale has a uniform distribution over [-1,1] to be consistent with the Russi paper and the methodologies in both Russi and Constantine.

```python
# QOI setup

# Number of samples
M = 1000

# The input matrix
a = np.random.uniform(-1,1,1000,30)

# An M x 1 column vector containing the evaluations of f at data points 1,...,M
def fun(a):
    
    a = a.copy()
    a = np.atleast_2d(a)
    M = a.shape[0]
    
    x = a[:,0]; y = a[:,1]; z = a[:,2]; w = a[:,3]; p = a[:,4];
    
    return (np.sin(a)+(y*z*p)+(a)*(w)**2+np.exp(z)*np.cos(p**3)).reshape(M, 1)

#The function's output
f = fun(a)

#Perturbed function output
pert_f=fun(a)+np.linalg.norm(fun(a))*0.001*np.random.rand(1000,1)

# A matrix of dimension M x 16. Row k is the transpose of the gradient at data point k.
#def grad_fun(a):

 #   a = a.copy()
  #  a = np.atleast_2d(a)
    
 #   x = a[:,0]; y = a[:,1]; z = a[:,2]; w = a[:,3]; p=a[:,4]; q=a[:,5]
    
  #  dfd1 = (2*A*x+2*A*y)[:,None]
  #  dfd2 = (2*A*x+2*A*y)[:,None]
  #  dfd3 = (2*B*z+2*B*w)[:,None]
  #  dfd4 = (2*B*z+2*B*w)[:,None]
  #  dfd5 = (2*C*p+2*C*q)[:,None]
  #  dfd6 = (2*C*p+2*C*q)[:,None]
  
  #  return np.hstack((dfd1, dfd2, dfd3, dfd4, dfd5, dfd6,np.zeros((1000,10))))

# Returns gradient matrix perturbed with noise proportional to the matrix norm of the gradient
#def noisy_grad_fun(a):

    #return grad_fun(a)+np.linalg.norm(grad_fun(a))*0.001*np.random.rand(1000,16)
```

# Active Subspace Analysis

"For demonstration we initially pretend we don't have the gradient of $f$. In this scenario, you have 3 options: use a global linear model to approximate gradients, use a global quadratic model to approximate gradients, or use local linear models to approximate gradients. We demonstrate each in turn." (Constantine)

```python
#Instantiate a subspace object
ss = ac.subspaces.Subspaces()

#Compute the subspace with a global linear model (sstype='OLS') and 100 bootstrap replicates
ss.compute(X=a, f=pert_f, nboot=100, sstype='OLS')
```

"We plot estimated eigenvalues, subspace errors, and a sufficient summary plot (a plot of the active variable vs data values)."

-Paul C.

```python
# GLOBAL LINEAR

#This plots the eigenvalues (ss.eigenvals) with bootstrap ranges (ss.e_br)
ac.utils.plotters.eigenvalues(ss.eigenvals, ss.e_br)

#This plots subspace errors with bootstrap ranges (all contained in ss.sub_br)
ac.utils.plotters.subspace_errors(ss.sub_br)

#This makes sufficient summary plots with the active variables (XX.dot(ss.W1)) and output (f)
ac.utils.plotters.sufficient_summary(a.dot(ss.W1), pert_f)

print(ss.eigenvecs)
print(a.dot(ss.W1))
```

"We now make the same plots with a global quadratic model (sstype='QPHD')." - Paul C.

```python
# GLOBAL QUADRATIC

ss.compute(X=a, f=pert_f, nboot=100, sstype='QPHD')
ac.utils.plotters.eigenvalues(ss.eigenvals, ss.e_br)
ac.utils.plotters.subspace_errors(ss.sub_br)
ac.utils.plotters.sufficient_summary(a.dot(ss.W1), pert_f)
print(ss.eigenvecs)
print(ss.W1)
```

"We now use local linear models to approximate gradients and make the same plots as above." - Paul C.

```python
# LOCAL LINEAR

df = ac.gradients.local_linear_gradients(a, pert_f) #Estimated gradients using local linear models

ss.compute(df=df, nboot=100)
ac.utils.plotters.eigenvalues(ss.eigenvals, ss.e_br)
ac.utils.plotters.subspace_errors(ss.sub_br)
ac.utils.plotters.sufficient_summary(a.dot(ss.W1), pert_f)
```

```python
# Analytical Gradients

df = noisy_grad_fun(a)
ss.compute(df=df, nboot=100)
ac.utils.plotters.eigenvalues(ss.eigenvals, ss.e_br)
ac.utils.plotters.subspace_errors(ss.sub_br)
ac.utils.plotters.sufficient_summary(a.dot(ss.W1), pert_f)

print ss.eigenvecs
```

```python
# Set degree of poylnomail approximation
RS = ac.utils.response_surfaces.PolynomialApproximation(1)

#Train the surface with active variable values and function values, pert_f
y = a.dot(ss.W1)
RS.train(y, pert_f)
print 'The R^2 value of the response surface is {:.4f}'.format(RS.Rsqr)

#Plot the data and response surface prediction
plt.figure(figsize=(7, 7))
y0 = np.linspace(-2, 2, 200)
plt.plot(y, f, 'bo', y0, RS.predict(y0[:,None])[0], 'k-', linewidth=2, markersize=8)
plt.grid(True)
plt.xlabel('Active Variable Value', fontsize=18)
plt.ylabel('Output', fontsize=18)
print(np.min(RS.predict(y0[:,None])[0]))
plt.show()
```

```python
# Set degree of poylnomail approximation
RS = ac.utils.response_surfaces.PolynomialApproximation(2)

#Train the surface with active variable values and function values, pert_f
y = a.dot(ss.W1)
RS.train(y, pert_f)
print 'The R^2 value of the response surface is {:.4f}'.format(RS.Rsqr)

#Plot the data and response surface prediction
plt.figure(figsize=(7, 7))
y0 = np.linspace(-2, 2, 200)
plt.plot(y, f, 'bo', y0, RS.predict(y0[:,None])[0], 'k-', linewidth=2, markersize=8)
plt.grid(True)
plt.xlabel('Active Variable Value', fontsize=18)
plt.ylabel('Output', fontsize=18)
print(np.min(RS.predict(y0[:,None])[0]))
plt.show()
```

```python
# Parameters alpha, beta, gamma from our function
A = 10000000 # alpha
B = 100000 # beta
C = 1000 # gamma

# An M x 1 column vector containing the evaluations of f at data points 1,...,M
def fun(a):
    
    a = a.copy()
    a = np.atleast_2d(a)
    M = a.shape[0]
    
    x = a[:,0]; y = a[:,1]; z = a[:,2]; w = a[:,3]; p = a[:,4]; q = a[:,5]; 
    
    return (A*(x+y)**2+B*(z+w)**2+C*(p+q)**2).reshape(M, 1)

#The function's output
f = fun(a)

#Perturbed function output
pert_f=fun(a)+np.linalg.norm(fun(a))*0.001*np.random.rand(1000,1)

# A matrix of dimension M x 16. Row k is the transpose of the gradient at data point k.
def grad_fun(a):

    a = a.copy()
    a = np.atleast_2d(a)
    
    x = a[:,0]; y = a[:,1]; z = a[:,2]; w = a[:,3]; p=a[:,4]; q=a[:,5]
    
    dfd1 = (2*A*x+2*A*y)[:,None]
    dfd2 = (2*A*x+2*A*y)[:,None]
    dfd3 = (2*B*z+2*B*w)[:,None]
    dfd4 = (2*B*z+2*B*w)[:,None]
    dfd5 = (2*C*p+2*C*q)[:,None]
    dfd6 = (2*C*p+2*C*q)[:,None]
  
    return np.hstack((dfd1, dfd2, dfd3, dfd4, dfd5, dfd6,np.zeros((1000,10))))

# Returns gradient matrix perturbed with noise proportional to the matrix norm of the gradient
def noisy_grad_fun(a):

    return grad_fun(a)+np.linalg.norm(grad_fun(a))*0.001*np.random.rand(1000,16)
```

```python
pca = PCA(n_components=3)
pca.fit(noisy_grad_fun(a))
print(pca.components_)
print(pca.explained_variance_ratio_)
```
